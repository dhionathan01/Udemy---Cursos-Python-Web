```python
file_open = open('notas_estudantes.txt')
file_content = file_open.read()
```

```python
des = file_content.split('\n')
print(des)
```

```python
lista_de_alunos = []
for i in des:
    lista_de_alunos.append(i.rstrip(':'))
print(lista_de_alunos)
```

```python
print (lista_de_alunos[0])
```

```python
lista_test = []
for x in lista_de_alunos:
    lista_test.append(x.split(':'))
print(lista_test)
```

```python
print(lista_test[0][0])
print(lista_test[0][1])
```

```python
notas = []
for x in lista_test:
    for i in range(len(x)-1):
        notas.append(int(x[i+1]))
            
print(notas)
```

```python
notas_aluno = []
for n in notas:
    notas_aluno.(n)
```
