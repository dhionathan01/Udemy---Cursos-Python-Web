```python
openFile = open('notas_estudantes.txt')
numeroLinhas = 0
arquivo = []
for fileLine in openFile:
    numeroLinhas = numeroLinhas + 1
    
openFile.seek(0)
for a in range(numeroLinhas):
    arquivo.append(openFile.readline().strip('\n'))
for b in range(numeroLinhas):
    arquivo[b] = arquivo[b].rstrip(':')
    arquivo[b] = arquivo[b].split(':')

for e in range(numeroLinhas):
    for f in range(1, len(arquivo[e])):
        arquivo[e][f] = int(arquivo[e][f])
for g in range(numeroLinhas):
    if sum(arquivo[g][1:])/(len(arquivo[g][1:])) >= 6:
        print(arquivo[g][0])
```

```python
for i in range(numeroLinhas):
    media = 0
    for j in range(len(arquivo[i][1:])):
        media = sum (arquivo[i][1:])/len(arquivo[i][1:])
        print("{:9}) - media: {:.2f}".format(arquivo[i][0], media))
```
